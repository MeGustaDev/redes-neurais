###  Estas são algumas das funções de ativação mais comuns, mas existem outras variações e funções personalizadas que podem ser usadas em casos específicos, dependendo da natureza da tarefa e da arquitetura da rede neural. A escolha da função de ativação pode ter um grande impacto no desempenho e no treinamento da rede, e geralmente é uma parte importante do projeto da arquitetura da rede neural.

```python
import numpy as np
```

```python
def sinal(self, x):
    return np.where(x >= 0.0, 1, -1)

def tanh(self, x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

def d_tanh(self, x):
    return 1 - self.tanh(x) * self.tanh(x)

def ReLU(self, x):
    return np.where(x > 0.0, x, 0.1*x)  #max(0, x) .... float

def d_ReLU(self, x):
    return np.where(x > 0.0, 1, 0.1)
```
